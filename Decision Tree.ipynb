```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
# another way of loading dataset
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
# Separating dependent and independent features
```

```python
X=df.iloc[:,:-1]
y=iris.target
```

```python
X,y
```

```python
# train test split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
   X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

### Post-Pruning technique


```python
treemodel=DecisionTreeClassifier()
treemodel.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel,filled=True)
```

```python
#lets define hyperparamter
treemodelh=DecisionTreeClassifier(max_depth=2)
```

```python
treemodelh.fit(X_train,y_train)
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel,filled=True)
```

```python
#prediction
y_pred=treemodelh.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
score=accuracy_score(y_pred,y_test)
score
```

```python
print(classification_report(y_pred,y_test))
```

## Pre-Pruning 

```python
parameter={'criterion':['gini','entropy','log_loss'],
          'splitter':['best','random'],
          'max_depth':[1,2,3,4,5],
          'max_features':['auto','sqrt','log2'],
           
          }
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
cv=GridSearchCV(treemodel,param_grid=parameter,cv=5,scoring='accuracy')
```

```python
cv.fit(X_train,y_train)
```

```python
print(cv.best_params_)
```

```python
print(cv.best_score_)
```

```python
y_predp=cv.predict(X_test)
```

```python
scorep=accuracy_score(y_predp,y_test)
scorep
```

```python
print(classification_report(y_predp,y_test))
```
